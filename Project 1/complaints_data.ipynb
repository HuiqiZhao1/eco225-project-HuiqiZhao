```python
import pandas as pd

df = pd.read_csv('consumer_complaints.csv')
df['product'].value_counts()
```

```python
student_loan_df = df[df['product'] == 'Student loan']
print("Number of student loan complaints:", len(student_loan_df))
```

```python
student_loan_df['sub_product'].value_counts()
```

```python
student_loan_df['issue'].value_counts()
```

```python
print("Earlist complaint date:", student_loan_df['date_received'].min())
print("Latest complaint date:", student_loan_df['date_received'].max())
```

```python
student_loan_df.to_csv('student_loan_complaints.csv', index=False)
```
