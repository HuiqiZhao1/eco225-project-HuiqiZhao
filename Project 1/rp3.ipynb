```python
import pandas as pd
import statsmodels.formula.api as smf
from stargazer.stargazer import Stargazer
from IPython.display import HTML

# --- 1. 数据加载与预处理 ---
df = pd.read_csv('final_data_cleaned.csv')

# 1.1 处理核心自变量 (X): Firm Type
# 将 'Specialist' 设为 1，'Bank' 设为 0
df['is_specialist'] = (df['institution_type'] == 'Specialist').astype(int)

# 1.2 处理机制变量 (Z): Resolution Strategy (Financial Relief)
# 将 'Relief' (有财务救济) 设为 1，其他 (Explanation, Other) 设为 0
df['has_relief'] = (df['company_response_to_consumer'] == 'Relief').astype(int)

# 1.3 处理因变量 (Y): 为了避免特殊字符报错，重命名列
# 'consumer_disputed?' -> 'disputed'
df = df.rename(columns={'consumer_disputed?': 'disputed'})

# --- 2. 建立回归模型 (按照 Week 8 要求增加复杂度) ---

# 模型 1: 基础回归 (X -> Y)
# 研究 Specialist 是否比 Bank 有更高的争议率
model1 = smf.ols('disputed ~ is_specialist', data=df).fit()

# 模型 2: 加入机制变量 (X + Z -> Y)
# 控制了是否给钱 (Relief) 后，Firm Type 的影响还在吗？
model2 = smf.ols('disputed ~ is_specialist + has_relief', data=df).fit()

# 模型 3: 加入控制变量 (X + Z + Controls -> Y)
# 加入资产规模 (sl_assets_billions) 和执法历史 (enforcement_action)
model3 = smf.ols('disputed ~ is_specialist + has_relief + sl_assets_billions + enforcement_action', data=df).fit()

# 模型 4: 交互项分析 (Interaction)
# Specialist 给钱的效果 (Relief) 是否和 Bank 给钱的效果不同？
model4 = smf.ols('disputed ~ is_specialist * has_relief + sl_assets_billions + enforcement_action', data=df).fit()

# --- 3. 生成学术表格 (Stargazer) ---
stargazer = Stargazer([model1, model2, model3, model4])

# 3.1 自定义表格标题和变量名（让报告看起来更专业）
stargazer.title('Regression Results: Effect of Firm Type on Dispute Rates')
stargazer.custom_columns(['Baseline', '+ Mechanism', '+ Controls', 'Interaction'], [1, 1, 1, 1])
stargazer.rename_covariates({
    'is_specialist': 'Specialist Servicer (vs. Bank)',
    'has_relief': 'Financial Relief Provided',
    'is_specialist:has_relief': 'Specialist × Relief (Interaction)',
    'sl_assets_billions': 'Asset Size ($B)',
    'enforcement_action': 'History of Enforcement'
})

# 3.2 隐藏不需要的统计信息（可选）
stargazer.show_model_numbers(False)  # 隐藏模型编号

# --- 4. 显示表格 ---
# 在 Jupyter 中直接渲染 HTML
HTML(stargazer.render_html())
```
