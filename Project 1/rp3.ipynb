```python
import pandas as pd
import statsmodels.formula.api as smf
from stargazer.stargazer import Stargazer
from IPython.display import HTML

# --- 1. 数据加载与预处理 ---
df = pd.read_csv('final_data_cleaned.csv')

# 1.1 处理核心自变量 (X): Firm Type
# 将 'Specialist' 设为 1，'Bank' 设为 0
df['is_specialist'] = (df['institution_type'] == 'Specialist').astype(int)

# 1.2 处理机制变量 (Z): Resolution Strategy (Financial Relief)
# 将 'Relief' (有财务救济) 设为 1，其他 (Explanation, Other) 设为 0
df['has_relief'] = (df['company_response_to_consumer'] == 'Relief').astype(int)

# 1.3 处理因变量 (Y): 为了避免特殊字符报错，重命名列
# 'consumer_disputed?' -> 'disputed'
df = df.rename(columns={'consumer_disputed?': 'disputed'})

# --- 2. 建立回归模型 (按照 Week 8 要求增加复杂度) ---

# 模型 1: 基础回归 (X -> Y)
# 研究 Specialist 是否比 Bank 有更高的争议率
model1 = smf.ols('disputed ~ is_specialist', data=df).fit()

# 模型 2: 加入机制变量 (X + Z -> Y)
# 控制了是否给钱 (Relief) 后，Firm Type 的影响还在吗？
model2 = smf.ols('disputed ~ is_specialist + has_relief', data=df).fit()

# 模型 3: 加入控制变量 (X + Z + Controls -> Y)
# 加入资产规模 (sl_assets_billions) 和执法历史 (enforcement_action)
model3 = smf.ols('disputed ~ is_specialist + has_relief + sl_assets_billions + enforcement_action', data=df).fit()

# 模型 4: 交互项分析 (Interaction)
# Specialist 给钱的效果 (Relief) 是否和 Bank 给钱的效果不同？
model4 = smf.ols('disputed ~ is_specialist * has_relief + sl_assets_billions + enforcement_action', data=df).fit()

# --- 3. 生成学术表格 (Stargazer) ---
stargazer = Stargazer([model1, model2, model3, model4])

# 3.1 自定义表格标题和变量名（让报告看起来更专业）
stargazer.title('Regression Results: Effect of Firm Type on Dispute Rates')
stargazer.custom_columns(['Baseline', '+ Mechanism', '+ Controls', 'Interaction'], [1, 1, 1, 1])
stargazer.rename_covariates({
    'is_specialist': 'Specialist Servicer (vs. Bank)',
    'has_relief': 'Financial Relief Provided',
    'is_specialist:has_relief': 'Specialist × Relief (Interaction)',
    'sl_assets_billions': 'Asset Size ($B)',
    'enforcement_action': 'History of Enforcement'
})

# 3.2 隐藏不需要的统计信息（可选）
stargazer.show_model_numbers(False)  # 隐藏模型编号

# --- 4. 显示表格 ---
# 在 Jupyter 中直接渲染 HTML
HTML(stargazer.render_html())
```

```python
import pandas as pd
import statsmodels.formula.api as smf
from stargazer.stargazer import Stargazer
from IPython.core.display import HTML

# 1. Load the cleaned dataset
df = pd.read_csv("final_data_cleaned.csv")

# 2. Variable Preparation
# Convert categorical variables into numeric/dummy variables for regression
df['is_relief'] = (df['company_response_to_consumer'] == 'Relief').astype(int)
df['is_bank'] = (df['institution_type'] == 'Bank').astype(int)
df['timely_binary'] = (df['timely_response'] == 'Yes').astype(int)
df['consumer_disputed'] = df['consumer_disputed?'].astype(float)

# 3. Build 4 Regression Models with increasing complexity [cite: 34, 35]
# Model 1: Basic relationship between relief and disputes
model1 = smf.ols('consumer_disputed ~ is_relief', data=df).fit()

# Model 2: Add firm type (Bank vs Specialist)
model2 = smf.ols('consumer_disputed ~ is_relief + is_bank', data=df).fit()

# Model 3: Add firm size (Assets)
model3 = smf.ols('consumer_disputed ~ is_relief + is_bank + sl_assets_billions', data=df).fit()

# Model 4: Full model including response timeliness
model4 = smf.ols('consumer_disputed ~ is_relief + is_bank + sl_assets_billions + timely_binary', data=df).fit()

# 4. Create the Stargazer table [cite: 37]
stargazer = Stargazer([model1, model2, model3, model4])

# Customize the table appearance
stargazer.title('Regression Results: Impact of Relief and Firm Type on Disputes')
stargazer.custom_columns(['Model 1', 'Model 2', 'Model 3', 'Model 4'], [1, 1, 1, 1])
stargazer.show_adj_r2 = True

# 5. Render the table as HTML for display [cite: 38]
# You can take a screenshot of this output for your PDF report
HTML(stargazer.render_html())
```

```python
import pandas as pd
import statsmodels.formula.api as smf
from stargazer.stargazer import Stargazer
from IPython.core.display import HTML

# 1. Load data
df = pd.read_csv("final_data_cleaned.csv")

# 2. Variable Prep
df['is_relief'] = (df['company_response_to_consumer'] == 'Relief').astype(int)
df['is_bank'] = (df['institution_type'] == 'Bank').astype(int)
df['timely_binary'] = (df['timely_response'] == 'Yes').astype(int)
df['consumer_disputed'] = df['consumer_disputed?'].astype(float)

# 3. Models (Model 4 now includes Issue Fixed Effects)
model1 = smf.ols('consumer_disputed ~ is_relief', data=df).fit()
model2 = smf.ols('consumer_disputed ~ is_relief + is_bank', data=df).fit()
model3 = smf.ols('consumer_disputed ~ is_relief + is_bank + sl_assets_billions', data=df).fit()
# Using C(issue) for fixed effects
model4 = smf.ols('consumer_disputed ~ is_relief + is_bank + sl_assets_billions + timely_binary + C(issue)', data=df).fit()

# 4. Create Table
stargazer = Stargazer([model1, model2, model3, model4])
stargazer.title('Regression Results')
stargazer.custom_columns(['Basic', 'Firm Type', 'Asset Size', 'Full (Issue FE)'], [1, 1, 1, 1])

# Omit the individual issue coefficients to keep the table readable
stargazer.add_custom_notes(['Model 4 includes Issue Fixed Effects.'])
# If you want to hide the many issue rows:
# stargazer.custom_grep_covariates('C\(issue\)') 

# 5. Render
HTML(stargazer.render_html())
```
